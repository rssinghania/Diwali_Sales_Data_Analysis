# Exploratory Data Analysis

```python
df.columns
```

```python
#sns.countplot(x = df['Gender'])
ax = sns.countplot(x = 'Gender', data = df )

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

# Age

```python
df.columns
```

```python
#sns.countplot(x = df['Age Group'], hue = df['Gender'])
ax = sns.countplot(x = 'Age Group', hue = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
print(sales_age)
```

From above graphs we can see that most of the buyers are of age group between 26-35 years Femails:

# State

```python
df.head()
```

```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)
sns.barplot(x = 'State', y = 'Orders', data = sales_state)
sns.set(rc = {'figure.figsize': (10,5)})
#print(sales_state.head(10))
```

```python
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(10)
sns.barplot(x = 'State', y = 'Amount', data = sales_state)
sns.set(rc = {'figure.figsize': (22,5)})
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Kernataka respectively

# Marital Status

```python
ax = sns.countplot(x ='Marital_Status', data = df)
sns.set(rc={'figure.figsize': (10,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
MS = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(x = 'Marital_Status', y = 'Amount', data = MS, hue = )
```

*From above graph we can see that most of the buyers are married (women) and they have high purchasing power*

# Occupation

```python
sns.set(rc={'figure.figsize': (20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.head(5)
```

```python
Occ = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(x = 'Occupation', y = 'Amount', data= Occ)
```

*From above graph we can see that most of the buyers are working in IT, Healthcare and Aviation sector:*

# Product Category

```python
ax = sns.countplot(x = 'Product_Category', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
pro_exp = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(10)

sns.barplot(x = 'Product_Category', y = 'Amount', data = pro_exp)
sns.set(rc={'figure.figsize':(20,5)})
```

*From above graph we can see that most of the sold products are from Food, Clothing and Electronics category:*

# Which product is more sale:

```python
pro = df.groupby(['Product_ID'], as_index=False)['Orders'].count().sort_values(by = 'Orders', ascending=False).head(10)

sns.barplot(x ='Product_ID', y = 'Orders', data = pro)
```
